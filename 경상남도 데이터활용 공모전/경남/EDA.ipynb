```python
for i in range(len(address)) :
    a = address[i].split(" ")
    address[i] = " ".join(a[0:6])

print(address)
```

```python
from geopy.geocoders import Nominatim
geo_local = Nominatim(user_agent='South Korea')

# 위도, 경도 반환하는 함수 생성
def geocoding(address):
    try:
        geo = geo_local.geocode(address)
        x_y = [geo.latitude, geo.longitude]
        return x_y

    except:
        return [0,0]
```

```python
latitude = []
longitude =[]

for i in address:
    latitude.append(geocoding(i)[0])
    longitude.append(geocoding(i)[1])
```

```python
address_df = pd.DataFrame({'주소':address,'위도':latitude,'경도':longitude})

address_df
```

```python
import csv

f = open("주민등록인구통계.csv", "r", encoding="utf-8")
age = csv.reader(f)
result = []
result2 = []
result3 = []
result4 = []
result5 = []

for row in age :
    if "의창구" in row[1] :
        for i in row[2:3] :
            result.append(int(i))
            
    if "성산구" in row[1] :
        for i in row[2:3] :
            result2.append(int(i))    
            
    if "마산합포구" in row[1] :
        for i in row[2:3] :
            result3.append(int(i))
            
    if "마산회원구" in row[1] :
        for i in row[2:3] :
            result4.append(int(i))  
            
    if "진해구" in row[1] :
        for i in row[2:3] :
            result5.append(int(i))
            
            
print("의창구 인구 수 :", sum(result))
print("성산구 인구 수 :", sum(result2))
print("마산합포구 인구 수 :", sum(result3))
print("마산회원구 인구 수 :", sum(result4))
print("진해구 인구 수 :", sum(result5))
```

# 변수 선택

```python
import pandas as pd

value = pd.read_csv("./변수 종합.csv")
value.head(5)
```

```python
import numpy as np
np.set_printoptions(precision=3)

# 쉼표 제거
value = value.replace(',', '', regex=True)
value.head(5)
```

```python
from sklearn.preprocessing import StandardScaler

# 1열을 제외한 데이터 선택
value_to_scale = value.iloc[:, 1:]

# 데이터프레임의 변수들을 스케일링
scaler = StandardScaler()
scaled_data = scaler.fit_transform(value_to_scale)
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=4, random_state=7)
pca.fit(scaled_data)
X_pca = pca.transform(scaled_data)
X_pca
```

```python
pca_df = pd.DataFrame(X_pca)
pca_df.head(5)
```

```python
result = pd.DataFrame({'설명가능한 분산 비율(고윳값)':pca.explained_variance_,'기여율': pca.explained_variance_ratio_},
                      index = np.array([f"pca{num}" for num in range(4)]))
result['누적기여율'] = result['기여율'].cumsum()
result
```

```python
# 변수들의 가중치 계수 확인
components = pca.components_
variables = pca_df.columns

# 결과 출력
for i, pc in enumerate(components):
    pc_weights = pd.Series(pc)
    print(f"PC{i} 가중치 계수:\n{pc_weights}\n")
```
