# 유통사

```python
# 구매일자를 날짜형 데이터로 변환
from datetime import datetime
df2["구매 일자"] = pd.to_datetime(df2["구매 일자"])

# 연월로 변경
df2["연월"] = df2["구매 일자"].dt.strftime("%Y%m")
df2.head()
```

```python
df2_month_1 = df2.groupby(["연월", "고객번호"], as_index = False)["구매 시간"].count()
df2_month_1
```

```python
# 월 별 고객 별 구매 횟수 데이터
df2_month_1.rename(columns = {"구매 시간" : "cnt"}, inplace = True)
df2_month_1
```

```python
df2_month_2 = df2_month_1.groupby(["고객번호"], as_index = False)["cnt"].agg([np.mean, np.median, np.max, np.min])
df2_month_2
```

```python
df2_month_2 = df2_month_2.reset_index(drop = False)
df2_month_2
```

```python
df2_join = pd.merge(df2, df2_month_2, on = "고객번호", how = "left")
df2_join.head()
```

```python
len(df2_join)
```

# 클러스터 별 특징

```python
df2_month = df2_join[["mean", "median", "amax", "amin"]]
df2_month
```

```python
# 표준화 작업
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
df2_month_scaled = sc.fit_transform(df2_month)
pd.DataFrame(df2_month_scaled)
```

```python
# 개수를 4로 지정해서 클러스터링
kmeans = KMeans(n_clusters = 4)
clusters = kmeans.fit(df2_month_scaled)

# cluster 값을 원본 데이터에 넣기
df2_month["clusters"] = clusters.labels_
df2_month
```

```python
df2_month.groupby("clusters").count()
```

```python
from sklearn.decomposition import PCA
X = df2_month_scaled.copy()

# 객체
pca = PCA(n_components = 2)

# 적용
pca.fit(X)
x_pca = pca.transform(X)
x_pca
```

```python
pca_df = pd.DataFrame(x_pca)
pca_df["cluster"] = df2_month["clusters"]
pca_df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
axs = plt.subplots()
axs = sns.scatterplot(0,1, hue = "cluster", data = pca_df)
```

```python
for i in sorted(pca_df["cluster"].unique()) :
  tmp = pca_df.loc[pca_df["cluster"] == i]
  plt.scatter(tmp[0], tmp[1])
  plt.legend(sorted(pca_df["cluster"].unique()))
```

```python
# 엘보우 방법을 통한 클러스터 개수 확인
inertia = []
for k in range(2, 7) :
  km = KMeans(n_clusters = k)
  km.fit(X)
  inertia.append(km.inertia_)
plt.plot(range(2,7), inertia)
plt.xlabel("k")
plt.ylabel("inertia")
plt.show()
```

```python
df2_join["clusters"] = df2_month["clusters"]
df2_join.head()
```

```python
df2_new = df2_join.groupby(["clusters", "score"], as_index = False).count()[["clusters", "score", "고객번호"]]
df2_new
```
