# 안전시설데이터

```python
df1 = pd.read_csv('/content/sample_data/안전시설 데이터.csv', encoding='utf-8')

col_name = ['번호', '카테고리', '담당센터명', '시설명', '주소', '전화번호', '제공서비스', '위도', '경도']
df1.columns = col_name
df1
```

```python
df1.isnull().sum()
```

```python
df1_1 = df1[df1["주소"].str.contains("서울특별시",na = False)]
df1_1
```

```python
safety_list = []

for i in df1_1["주소"] :
  safety = i.split()[1]
  safety_list.append(safety)

df1_1["시군구명"] = safety_list
df1_1
```

```python
df1_1 = df1_1[["번호", "담당센터명", "시설명", "위도", "경도", "시군구명"]]
df1_1
```

```python
df1_1.pivot_table(index = "시군구명", values = "번호", aggfunc = "count").rename(columns = {"번호" : "안전시설 개수"})
```

```python
df1_1.pivot_table(index = "담당센터명", values = "번호", aggfunc = "count").rename(columns = {"번호" : "안전시설 개수"})
```

```python
import folium
import json
```

```python
safety_map = folium.Map(
    location = [37.573050, 126.979189],
    tiles = "Stamen Terrain",
    zoom_start = 11
)
safety_map
```

```python
for i in df1_1.index :
  lat = df1_1.loc[i, "위도"]
  lng = df1_1.loc[i, "경도"]

  folium.CircleMarker(
      location = [lat, lng],
      fill = True,
      fill_color = "green",
      fill_opacity = 1,
      color = "yellow",
      weight = 1,
      radius = 3
  ).add_to(safety_map)

safety_map
```
