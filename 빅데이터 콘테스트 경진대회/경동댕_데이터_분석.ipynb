```python
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor

X_train = data[["CCTV 개수", "안전시설 개수", "여성안심지킴이집 개수", "심야시간 112 신고건수", "유동인구 수", "유흥업소 개수", "상주인구", "검거"]]

def feature_engineering_xbyVIF(X_train) :
  vif = pd.DataFrame()
  vif["VIF_Factor"] = [variance_inflation_factor(X_train.values, i)
  for i in range(X_train.shape[1])]
  vif["Feature"] = X_train.columns
  return vif
vif = feature_engineering_xbyVIF(X_train)
print(vif)
```

# 계층적 군집화

```python
data_x = data.loc[:, ["CCTV 개수", "여성안심지킴이집 개수", "안전시설 개수", "유흥업소 개수"]]
data_y = data.loc[:, ["범죄 발생수"]]
```

```python
#정규화
from sklearn.preprocessing import MinMaxScaler

data_scaler = MinMaxScaler()
data_x = data_scaler.fit_transform(data_x)

target_scaler = MinMaxScaler()
data_y = target_scaler.fit_transform(data_y)
```

```python
df = pd.DataFrame(data_x)
df
```

```python
from scipy.spatial.distance import pdist, squareform
Y = pdist(df)
Y
```

```python
row_dist = pd.DataFrame(squareform(pdist(df, metric = "euclidean")))
row_dist
```

```python
from scipy.cluster.hierarchy import linkage

row_clusters = linkage(pdist(df, metric = "euclidean"), method = "average")

row_clusters = linkage(df.values, metric = "euclidean", method = "average")

pd.DataFrame(row_clusters, index = ["cluster %d"%(i+1) for i in range(row_clusters.shape[0])])
```

```python
# 덴드로그램
import scipy.cluster.hierarchy as sch
import matplotlib.pyplot as plt

row_dendr = sch.dendrogram(row_clusters)

plt.tight_layout()
plt.ylabel("Euclidean distance")
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import silhouette_score

k_range = range(2,14)
for k in k_range :
  average_clustering = AgglomerativeClustering(n_clusters = k, linkage = "average")
  average_cluster = average_clustering.fit_predict(df)
  score = silhouette_score(df, average_cluster)
  print(score)
```

```python
from sklearn.cluster import AgglomerativeClustering as AC
clusters = AC(n_clusters = 4,
              linkage = "average")
clusters.fit(df)
df["지역유형"] = clusters.labels_
df
```
