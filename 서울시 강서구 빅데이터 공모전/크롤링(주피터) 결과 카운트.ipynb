```python
from konlpy.tag import Twitter
twitter = Twitter()
```

```python
type(txt)
```

```python
morph_list = []
for i in txt :
  morphs = twitter.morphs(i)
  for morph in morphs :
    morph_list.append(morph)
```

```python
morph_list
```

```python
len(morph_list)
```

```python
from collections import Counter
count = Counter(morph_list)
```

```python
count.most_common()
```

```python
for i in morph_list :
  f = open("morph_txt", "a", encoding="utf8")
  f.write(i + "\n")
  f.close()
```

# 크롤링(주피터에서 함)

```python
import requests
from bs4 import BeautifulSoup

# 1. 크롤링할 블로그 URL 결정
base_url = 'https://search.naver.com/search.naver?query='

keyword = input("검색어를 입력하세요 : ")
search_url = base_url + keyword
r = requests.get(search_url)

soup = BeautifulSoup(r.text, "html.parser")
items = soup.select(".api_txt_lines.total_tit")

for e, item in enumerate(items,1) :
  print(f"{e} : {item.text}")
```
